```python

import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
#plt.style.use("seaborn")

# please check which style are available
#print(plt.style.available)
```

```python
tickers = ['aapl', 'goog', 'meta', 'nvda']
```

```python
df = yf.download(tickers, period='3mo')

adj_close = df.loc[:, "Adj Close"].copy()

norm = adj_close.div(adj_close.iloc[0]).mul(100)

norm.plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```
