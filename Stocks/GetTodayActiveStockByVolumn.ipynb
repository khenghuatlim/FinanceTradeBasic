# The code will retrieve active stocks by volumn, it goes to Yahoo Finance and retrieve it

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#get first 100 active tickers by volumns
url_active  = "https://finance.yahoo.com/most-active/?offset=0&count=100"
```

```python
def getActiveStock():
    df = pd.DataFrame()
    data = requests.get(url_active, headers={"Content-Type":"text/html; charset=utf-8", 
                                             "Content-Encoding": "gzip",
                                             "Connection": "keep-alive",
                                             "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36 Edg/120.0.0.0"}).text
    
    soup = BeautifulSoup(data, 'html.parser') #'html.parser' , 'lxml' , 'html5lib'
    table = soup.find('table')

    # preparing the dataframe headers
    for thead in table.thead.tr.find_all('th'):
        if not "52" in thead.text.strip():
            df[thead.text.strip()] = thead.text.strip()



    for row in table.tbody.find_all('tr'):
        columns = row.find_all('td')    
        symbol  = columns[0].a.text.strip()
        name = columns[1].text.strip()
        price = columns[2].text.strip()
        change = columns[3].text.strip()
        change_percent = columns[4].text.strip()
        volumn  = columns[5].text.strip()
        avg_vol_3m = columns[6].text.strip()
        market_cap = columns[7].text.strip()
        pe_ratio = columns[8].text.strip()
        df.loc[len(df.index)] = [symbol, name, price, change, change_percent, volumn, avg_vol_3m, market_cap, pe_ratio]

    return df
```

```python
df =  getActiveStock()
```
